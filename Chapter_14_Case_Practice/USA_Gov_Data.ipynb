```python
## Seperate Data by operation system. If windows appear in a then window
```

```python
#cframe = frame[]
cframe = frame[frame['a'].notnull()]
cframe.head(10)
```

```python
cframe['os'] = cframe['a'].apply(lambda x: 'Windows' if 'Windows' in str(x) else 'Not Windows')
cframe.head(10)
```

```python
by_tz_os = cframe.groupby(['tz', 'os'])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts.head(5)
```

```python
agg_counts_2 = by_tz_os.size()
agg_counts_2.head(10)
```
