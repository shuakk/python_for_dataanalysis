```python
data_file_dir = '/Users/KK/Documents/python_for_dataanalysis/data/group_by_practice.csv'
df = pd.read_csv(data_file_dir, skiprows = [0], header=None)
df.head(5)
```

```python
df.columns = ['team', 'rep_id', 'month', 'sales_amount']
df.head(5)
```

## Add column based on logic calculation

```python
df['sales_amount_category'] = df['sales_amount'].apply(lambda x: 'low amount' if x<20 else 'normal')
df.head(5)
```

```python
def sale_amount_detail_category(x):
    result = str()
    if x < 20:
        result = 'low amount'
    elif x < 30:
        result = 'normal amount'
    else:
        result = 'excellent amount'
    return result

df['sales_amount_detail_category'] = df['sales_amount'].apply(lambda x: sale_amount_detail_category(x))
df.head(5)
```

## Group By and Join

```python
df_jan = df[df['month'] == '2018-01']
df_jan.sort_values(by = 'team')
```

```python
df_jan['sales_amount_rmb'] = df_jan['sales_amount'] / 7
df_jan
```

```python
df_jan_groupby = df_jan.groupby(['team'])
df_jan_groupby.size()
```

```python
df_jan_groupby.sum()
```

```python
df_jan.groupby(['team'])[['team', 'sales_amount_rmb']].sum()
```
